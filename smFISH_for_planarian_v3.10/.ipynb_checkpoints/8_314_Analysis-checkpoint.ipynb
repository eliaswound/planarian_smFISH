```python
import pandas as pd
from scipy import stats

# Load the Excel file
file_path = "306_injury_site_length.xlsx"
df = pd.read_excel(file_path)

# Merge data for each time point
incision_0hr = np.concatenate(Incision_0h)
amputation_0hr = np.concatenate(Amputation_0h)

incision_6hr = np.concatenate(Incision_6h)
amputation_6hr = np.concatenate(Amputation_6h)

incision_12hr = np.concatenate(Incision_12h)
amputation_12hr = np.concatenate(Amputation_12h)

# Perform independent t-tests
t_stat_0hr, p_value_0hr = stats.ttest_ind(incision_0hr, amputation_0hr, equal_var=False)
t_stat_6hr, p_value_6hr = stats.ttest_ind(incision_6hr, amputation_6hr, equal_var=False)
t_stat_12hr, p_value_12hr = stats.ttest_ind(incision_12hr, amputation_12hr, equal_var=False)

alpha = 0.05
print("\nStatistical Significance:")
print("0hr:", "Significant difference" if p_value_0hr < alpha else "No significant difference")
print(p_value_0hr)
print("6hr:", "Significant difference" if p_value_6hr < alpha else "No significant difference")
print(p_value_6hr)
print("12hr:", "Significant difference" if p_value_12hr < alpha else "No significant difference")
print(p_value_12hr)
```

```python
def compute_sums(data_arrays, first_value):
    return [arr.sum() / first_value for arr in data_arrays]

# Function to compute lengths divided by first number in dataframe
def compute_lengths(data_arrays, first_value):
    return [len(arr) / first_value for arr in data_arrays]

# Create new DataFrames
sum_df = pd.DataFrame({
    "Incision_0h": compute_sums(Incision_0h, df["0h_Incision"][0]),
    "Amputation_0h": compute_sums(Amputation_0h, df["0h_Amputation"][0]),
    "Incision_6h": compute_sums(Incision_6h, df["6h_Incision"][0]),
    "Amputation_6h": compute_sums(Amputation_6h, df["6h_Amputation"][0]),
    "Incision_12h": compute_sums(Incision_12h, df["12h_Incison"][0]),
    "Amputation_12h": compute_sums(Amputation_12h, df["12h_Amputation"][0])
})

length_df = pd.DataFrame({
    "Incision_0h": compute_lengths(Incision_0h, df["0h_Incision"][0]),
    "Amputation_0h": compute_lengths(Amputation_0h, df["0h_Amputation"][0]),
    "Incision_6h": compute_lengths(Incision_6h, df["6h_Incision"][0]),
    "Amputation_6h": compute_lengths(Amputation_6h, df["6h_Amputation"][0]),
    "Incision_12h": compute_lengths(Incision_12h, df["12h_Incison"][0]),
    "Amputation_12h": compute_lengths(Amputation_12h, df["12h_Amputation"][0])
})

# Display results
print("Average spot count per injury site length DataFrame:")
print(sum_df)
print("\nAverage cell expressing per injury site length:")
print(length_df)
```
