```python
from scipy.stats import ks_2samp
import pandas as pd

# KS test helper
def ks_test_by_condition(df, cond1, cond2):
    group1 = df[df["Condition"].str.startswith(cond1)]["Value"].dropna()
    group2 = df[df["Condition"].str.startswith(cond2)]["Value"].dropna()
    
    if len(group1) == 0 or len(group2) == 0:
        print(f" One or both groups are empty: {cond1} vs {cond2}")
        return None
    
    stat, p_value = ks_2samp(group1, group2)
    print(f"{cond1} vs {cond2} --> KS stat: {stat:.4f}, p-value: {p_value:.4e}")
    return stat, p_value

# All desired comparisons
comparisons = [
    ("Incision_12h", "Amputation_12h"),
    ("Incision_0h", "Amputation_0h"),
    ("Amputation_12h", "Incision_6h"),
    ("Incision_12h", "Incision_6h"),
    ("Incision_6h", "Amputation_6h")  # New comparison
]

# Run and collect results
results = []
for cond1, cond2 in comparisons:
    result = ks_test_by_condition(df_long, cond1, cond2)
    if result:
        stat, pval = result
        results.append({
            "Comparison": f"{cond1} vs {cond2}",
            "KS_stat": stat,
            "p_value": pval
        })

# Convert to DataFrame and sort by p-value
results_df = pd.DataFrame(results).sort_values(by="p_value")
print("\n Sorted KS Test Results (by p-value):")
print(results_df.to_string(index=False))
results_df.to_csv("/Volumes/Backup Plus/Experiment_results/314_analysis_results/Experiment/ks_test_results.csv", index=False)
```

```python
from scipy.stats import mannwhitneyu
def mann_whitney_by_condition(df, cond1, cond2, alternative="two-sided"):
    group1 = df[df["Condition"].str.startswith(cond1)]["Value"].dropna()
    group2 = df[df["Condition"].str.startswith(cond2)]["Value"].dropna()
    
    if len(group1) == 0 or len(group2) == 0:
        print(f" One or both groups are empty: {cond1} vs {cond2}")
        return None
    
    stat, p_value = mannwhitneyu(group1, group2, alternative=alternative)
    print(f"{cond1} vs {cond2} --> U stat: {stat:.4f}, p-value: {p_value:.4e}")
    return stat, p_value
comparisons = [
    ("Incision_12h", "Amputation_12h"),
    ("Incision_0h", "Amputation_0h"),
    ("Amputation_6h", "Amputation_12h"),
    ("Incision_6h", "Incision_12h"),
    ("Incision_6h", "Amputation_6h")  # Added one
]

# Store results
mw_results = []
for cond1, cond2 in comparisons:
    result = mann_whitney_by_condition(df_long, cond1, cond2, alternative="less")
    if result:
        stat, pval = result
        mw_results.append({
            "Comparison": f"{cond1} vs {cond2}",
            "U_stat": stat,
            "p_value": pval
        })

# Create DataFrame and sort
mw_results_df = pd.DataFrame(mw_results).sort_values(by="p_value")
print("\n MannWhitney U Test Results (sorted by p-value):")
print(mw_results_df.to_string(index=False))
mw_results_df.to_csv("mann_whitney_results.csv", index=False)
```
